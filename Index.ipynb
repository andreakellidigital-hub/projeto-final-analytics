```python
# Example of updating prices using a for loop and if/else
produtos = [
    {"nome": "Produto A", "preco": 100},
    {"nome": "Produto B", "preco": 250},
    {"nome": "Produto C", "preco": 50}
]
print("Pre√ßos originais:")
for produto in produtos:
    print(f"{produto['nome']}: R${produto['preco']:.2f}")
# Apply price updates
for produto in produtos:
    if produto["preco"] < 100:
        produto["preco"] *= 1.15  # Increase price by 15% for products under R$100
    elif produto["preco"] >= 100 and produto["preco"] < 200:
        produto["preco"] *= 1.10  # Increase price by 10% for products between R$100 and R$200
    else:
        produto["preco"] *= 1.05  # Increase price by 5% for products R$200 or more
print("\nPre√ßos atualizados:")
for produto in produtos:
    print(f"{produto['nome']}: R${produto['preco']:.2f}")
```

#<font color="blue"> 9. Estat√≠stica</font>

#### C√°lculo da M√©dia

<hr style="border: 1px solid rgba(242, 248, 241, 1); width: 100%; margin: 5px 0;">

A f√≥rmula da m√©dia √© usada para calcular o valor m√©dio de um conjunto de dados. A f√≥rmula √© dada por:

$$

\text{M√©dia} = \frac{\sum_{i=1}^{n} x_i}{n}

$$

Onde:

- $( x_i )$ representa cada valor do conjunto de dados.

- $( n )$ √© o n√∫mero total de elementos.

### Exemplo:

Se tivermos os seguintes n√∫meros: 2, 4, 6, 8, 10, a m√©dia ser√°:

$$

\text{M√©dia} = \frac{2 + 4 + 6 + 8 + 10}{5} = \frac{30}{5} = 6

$$

```python
# M√©dia

a = [2, 4, 6, 8, 10]

print(f'M√©dia: {sum(a) / len(a)}')
```

```python
import statistics as st

print(f'M√©dia: {st.mean(a)}')
```

```python
# n√∫mero √≠mpar de elementos

b = [1500, 1800, 2000, 2100, 2200, 2400]

print(f'Mediana: {sorted(b)[len(b) // 2] if len(b) % 2 != 0 else (sorted(b)[len(b) // 2 - 1] + sorted(b)[len(b) // 2]) / 2}')
```

```python
import numpy as np

import pandas as pd

from statistics import mode

notas = [8, 7, 6, 9, 7, 8, 10, 9, 6, 7, 8, 2, 10, 9, 7, 6, 8, 9, 5, 10]

media = np.mean(notas)

mediana = np.median(notas)

moda = mode(notas)

print("M√©dia   :", media)

print("Mediana :", mediana)

print("Moda    :", moda)
```

```python
amplitude = max(notas) - min(notas)

variancia = np.var(notas, ddof=1)   # ddof=1 ‚Üí amostral

desvio_padrao = np.std(notas, ddof=1)

cv = desvio_padrao / media

print("Amplitude        :", amplitude)

print("Vari√¢ncia        :", variancia)

print("Desvio Padr√£o    :", desvio_padrao)

print("Coef. Varia√ß√£o   :", cv)
```

#<font color="blue"> 10. Estrat√©gia</font>

Ela orienta o direcionamento das a√ß√µes e garante que todas as decis√µes estejam alinhadas aos resultados esperados.

A estrat√©gia √© o ponto de partida da an√°lise de dados para tomada de decis√£o.
Envolve:

**Definir objetivos claros** (o que queremos alcan√ßar).

**Estabelecer indicadores de desempenho** (KPIs) para medir progresso.

**Usar dados para embasar planos e evitar decis√µes apenas intuitivas.**

**##Exemplo pr√°tico no Colab**

Simula√ß√£o de uma empresa que quer aumentar as vendas em 20%:

```python
import pandas as pd

# Exemplo de dados de vendas (em milhares)
dados = {
    "M√™s": ["Jan", "Fev", "Mar", "Abr", "Mai", "Jun"],
    "Vendas": [100, 120, 130, 115, 140, 160]
}

df = pd.DataFrame(dados)

# Meta estrat√©gica: aumentar vendas em 20%
meta = df["Vendas"].mean() * 1.20

print("M√©dia atual de vendas:", df["Vendas"].mean())
print("Meta estrat√©gica (20% a mais):", meta)

# Monitoramento simples
df["Meta Atingida"] = df["Vendas"] >= meta
df
```

##üìä**Ciclo Estrat√©gico**

# **Estrat√©gia**
A estrat√©gia define a vis√£o e as metas principais.  
Exemplo: aumentar as vendas em **20%** no pr√≥ximo semestre.  
Isso ser√° a refer√™ncia para todas as an√°lises e decis√µes.


import pandas as pd
#**Plano**

# Dados fict√≠cios de vendas mensais (em milhares)
dados = {
    "M√™s": ["Jan", "Fev", "Mar", "Abr", "Mai", "Jun"],
    "Vendas": [100, 120, 130, 115, 140, 160]
}

df = pd.DataFrame(dados)

# Meta estrat√©gica: +20% sobre a m√©dia atual
meta = df["Vendas"].mean() * 1.20
print("Meta estrat√©gica:", round(meta, 2))
df


#**Monitoramento / An√°lise**
# Verifica m√™s a m√™s se a meta foi atingida
df["Meta Atingida"] = df["Vendas"] >= meta
df

# Percentual de meses que atingiram a meta
atingidos = df["Meta Atingida"].mean() * 100
print(f"Meta atingida em {atingidos:.1f}% dos meses")


# **Atua√ß√£o / Ajuste**
Se a meta n√£o est√° sendo atingida, sugerir ajustes

```python
# Percentual de meses que atingiram a meta
atingidos = df["Meta Atingida"].mean() * 100
print(f"Meta atingida em {atingidos:.1f}% dos meses")
```

#üìä **Visualiza√ß√£o: Vendas vs Meta**

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.plot(df["M√™s"], df["Vendas"], marker="o", label="Vendas")
plt.axhline(y=meta, color="r", linestyle="--", label=f"Meta ({round(meta,1)})")

plt.title("Vendas Mensais x Meta Estrat√©gica")
plt.xlabel("M√™s")
plt.ylabel("Vendas (milhares)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

# **Conclus√£o do Ciclo Estrat√©gico**

```python
# Estudo de Caso: Analisando Vendas Semanais
import pandas as pd
import matplotlib.pyplot as plt

# Dados fict√≠cios de vendas (exemplo do dia a dia)
dados = {
    "Dia": ["Seg", "Ter", "Qua", "Qui", "Sex", "S√°b", "Dom"],
    "Vendas": [20, 35, 30, 25, 50, 70, 40]
}

# Criando DataFrame
df = pd.DataFrame(dados)

# Exibir tabela
print("Tabela de Vendas Semanais:")
print(df)

# Gerar gr√°fico
plt.figure(figsize=(8,5))
plt.plot(df["Dia"], df["Vendas"], marker="o")
plt.title("Vendas Semanais")
plt.xlabel("Dia da Semana")
plt.ylabel("Quantidade Vendida")
plt.grid(True)
plt.show()
```

```python
# Estudo de Caso: Analisando Engajamento no Instagram
import pandas as pd
import matplotlib.pyplot as plt

# Dados fict√≠cios de posts (exemplo pr√°tico do dia a dia)
dados = {
    "Post": ["Post 1", "Post 2", "Post 3", "Post 4", "Post 5"],
    "Curtidas": [120, 340, 230, 150, 400],
    "Comentarios": [15, 40, 25, 10, 60]
}

# Criando DataFrame
df = pd.DataFrame(dados)

# Exibir tabela
print("Tabela de Engajamento dos Posts:")
print(df)

# Gr√°fico de barras
df.plot(x="Post", y=["Curtidas", "Comentarios"], kind="bar", figsize=(8,5))
plt.title("Engajamento no Instagram")
plt.xlabel("Post")
plt.ylabel("Quantidade")
plt.show()
```
